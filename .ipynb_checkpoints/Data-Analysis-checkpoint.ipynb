```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Data-Analysis.csv')
df
```

# 1. Show a breakdown of distance from home by job role and attrition.

```python
df.groupby(['JobRole','Attrition']).DistanceFromHome.mean()
```

# 2. Compare average monthly income by education and attrition.

```python
attrition_monthlyrate = df.groupby(['Attrition']).MonthlyRate.mean()
attrition_monthlyrate
```

```python
education_monthlyrate = df.groupby(['Education']).MonthlyRate.mean()
education_monthlyrate
```

```python
sns.jointplot(data=df, x=attrition_monthlyrate, y=education_monthlyrate)
```
